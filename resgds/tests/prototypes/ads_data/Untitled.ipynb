```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_table('mirror_mismatch.txt',header=1,names=['freq','s21'])

freq = np.array(df['freq'].values)
s21 = np.array(df['s21'].values)

# plt.plot(freq,s21)
# plt.show()

# df.plot('freq','s21')
plt.plot(freq/1e09,s21)
plt.grid()
plt.xlabel('Frequency (GHz)',fontsize=22)
plt.ylabel('$S_{21}$ (dB)',fontsize=22)
plt.savefig('mirror_mismatch_ideal_18_121.jpg')
plt.tight_layout()
plt.show()
```
